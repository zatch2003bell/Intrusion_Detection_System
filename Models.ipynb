# Model

```python
# Importing Libraries

from sklearn.ensemble import AdaBoostClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

### 1. Ada Boast Algorithm

```python
# Fitting and Training 
adaboost_model = AdaBoostClassifier(n_estimators=50, random_state=42)
adaboost_model.fit(x_train, y_train.ravel())
```

```python
# Prediction
adaboost_predict = adaboost_model.predict(x_test)
```

```python
# Result
adaboost_accuracy = accuracy_score(y_test, adaboost_predict)
print("AdaBoost Classifier Accuracy:", adaboost_accuracy)
```

### 2. KNN Algorithm

```python
# Fitting and Training 
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train, y_train.ravel())
```

```python
# Prediction
knn_predict = knn_model.predict(x_test)
```

```python
# Result
knn_accuracy = accuracy_score(y_test, knn_predict)
print("KNN Classifier Accuracy:", knn_accuracy)
```

### 3. Decision Tree Algorithm

```python
# Fitting and Training 
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit((x_train, y_train.ravel())
```

```python
# Prediction
dt_predict = dt_model.predict(x_test)
```

```python
# Result
dt_accuracy = accuracy_score(y_test, dt_predict)
print("Decision Tree Classifier Accuracy:", dt_accuracy)
```
